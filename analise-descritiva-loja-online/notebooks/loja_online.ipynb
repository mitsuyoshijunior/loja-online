```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib.ticker import MultipleLocator
import matplotlib.dates as mdates
```

## 15 clientes que mais gastaram

```python
caminho_dados1 = '/content/drive/MyDrive/loja-online/15_clientes_mais_gastaram.csv'
```

```python
df1 = pd.read_csv(caminho_dados1)
```

```python
df1.head()
```

```python
# Configurando a figura, eixos e grade
fig, ax = plt.subplots(figsize=(18, 8))
sns.set_style("whitegrid")

# Configurando dados do gráfico
bars = ax.barh(df1['Cliente'], df1['Preco_total'], color="lightgreen")
ax.invert_yaxis()
ax.bar_label(bars, labels=[f'R${v:,.2f}' for v in df1['Preco_total']], padding=3, fontsize=10, color='white')


# Ajustando cores e legendas do gráfico
ax.set_facecolor('#3b3b47')
ax.set_title('TOP 15 clientes que mais gastaram até hoje', fontsize=16)
ax.grid(color='gray', linestyle='--', linewidth=0.5)
ax.set_ylabel('Clientes', fontsize = 12)
ax.set_xlabel('Valor em R$', fontsize = 12)
ax.xaxis.set_major_locator(MultipleLocator(5000))
ax.set_xlim(right=55000)

# Exibindo o gráfico
plt.show()
```

## Ranking das 20 cidades que mais compraram (em R$)

```python
caminho_dados3 = '/content/drive/MyDrive/loja-online/20_cidades_mais_compraram_antes_natal.csv'
```

```python
df3 = pd.read_csv(caminho_dados3)
```

```python
df3.head()
```

```python
# Configurando a figura, eixos e grade
fig, ax = plt.subplots(figsize=(16, 8))
ax.grid(color='lightgray', linestyle='--', linewidth=0.1)

# Criando iteração de cores para as 3 primeiras cidades
cores =[]
for idx, i in df3['Cidade'].items():
  if idx == 0:
    cores.append('gold')
  elif idx == 1:
    cores.append('silver')
  elif idx == 2:
    cores.append('#CD7F32')
  else:
    cores.append('lightgreen')

# Configurando os dados do gráfico
bars = ax.barh(df3['Cidade'] + ' - ' + df3['UF'], df3['Valor_gasto'], color=cores)
ax.bar_label(bars, labels=[f'R${v:,.2f}' for v in df3['Valor_gasto']], padding=3, fontsize=9, color='white')

# Ajustando cores e legendas
ax.set_facecolor('#3b3b47')
ax.invert_yaxis()
ax.set_title('Cidades que mais gastaram até hoje', fontsize=16)
ax.set_ylabel('Cidades', fontsize=10)
ax.set_xlabel('Valor em R$', fontsize=10)
ax.set_xlim(right=20000)




# Exibindo o gráfico
plt.show()
```

## Quantidade de produtos vendidos ao longo de 2024

```python
caminho_dados4 = '/content/drive/MyDrive/loja-online/quantidade_produtos_vendidos_2024.csv'
```

```python
df4 = pd.read_csv(caminho_dados4)
df4.head()
```

```python
df4['MesAno'] = pd.to_datetime(df4['MesAno'], format= '%m-%Y')
```

```python
df4.info()
```

```python
from datetime import date
# Configurando figura, eixos e grade do gráfico
fig, ax = plt.subplots(figsize=(16, 8))
ax.grid(color='lightgray', linestyle='--', linewidth=0.1)

# Configurando dados do gráfico
ax.plot(df4['MesAno'], df4['Quantidade'], color = 'lightgreen', marker = 'o', linewidth = 2)

for meses, quantidade in zip(df4['MesAno'], df4['Quantidade']):
  ax.text(meses, quantidade + 3, f'{quantidade}', ha = 'right', va = 'bottom', color = 'white')


# Ajustando cores e legendas
ax.set_facecolor('#3b3b47')
ax.set_title('Quantidade de Produtos Vendidos ao Longo de 2024', fontsize = 16)
ax.set_ylabel('Quantidade', fontsize = 12)
ax.set_xlabel('Meses', fontsize = 12)

# Configurando exibição dos ticks do eixo x
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b-%y'))

# Exibindo o gráfico
plt.show()
```

## Porcentagem de vendas por categoria até hoje

```python
caminho_dados5 = '/content/drive/MyDrive/loja-online/quantidade_vendas_por_categoria.csv'
```

```python
df5 = pd.read_csv(caminho_dados5)
```

```python
df5.head()
```

```python
df5.info()
```

```python
# Criando figura e eixo
fig, ax = plt.subplots(figsize=(16, 8))

# Configurando dados do gráfico de pizza
ax.pie(df5['Quantidade'], labels=df5['Categoria'],
        autopct='%1.1f%%',
        startangle=90,
        textprops={'fontsize':14, 'fontfamily':'Arial'})

# Ajustando o visual do gráfico
ax.axis('equal')
ax.set_title('Porcentagem de Venda por Categoria', fontsize = 18, pad = 20)

# Ajustando os elementos da figura para não sobrepor
plt.tight_layout()

# Exibindo o gráfico
plt.show()
```
